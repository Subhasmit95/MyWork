```python
#All the imports go here
import numpy as np
import cv2
import time
import cvzone
import matplotlib.pyplot as plt

from cvzone.FaceMeshModule import FaceMeshDetector
from cvzone.PlotModule import LivePlot
```

```python


count=0


#Starting the video capture
cap = cv2.VideoCapture(0)
ret,img = cap.read()
facedetector=FaceMeshDetector(maxFaces=1) #since we have only the patient face.
ids=[21,22,23,25,159,130,243]
PlotY=LivePlot(640,360,[20,30])
while(ret):
    ret,img = cap.read()
    faceimg,face=facedetector.findFaceMesh(img,draw=True)
    
    if face:
        facepoint=face[0]
#         for id in ids:
            
        leftup=facepoint[159]
        leftdown=facepoint[23]
        leftleft=facepoint[130]
        leftright=facepoint[243]
        cv2.circle(img,leftdown,2,(0,0,255))
        cv2.circle(img,leftup,2,(0,0,255))
        cv2.circle(img,leftleft,2,(0,0,255))
        cv2.circle(img,leftright,2,(0,0,255))
        leftlengthver, _ =facedetector.findDistance(leftup,leftdown)
        leftlengthhor, _ =facedetector.findDistance(leftleft,leftright)
        
        print(leftlengthver/leftlengthhor)
        
        #here we are plotting the live graph of the blink to check the threshold value for the main project
        imgplot=PlotY.update((leftlengthver/leftlengthhor)*100)
        cv2.imshow('Captureimage1',imgplot)
    
    cv2.imshow('Captureimage',img)
    a = cv2.waitKey(25)

    if(a==ord('q')):
        break
    elif(a==ord('s') and first_read):
        #This will start the detection
        first_read = False

cap.release()
cv2.destroyAllWindows()
```
